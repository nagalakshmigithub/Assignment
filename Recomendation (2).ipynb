```python
book =book.sort_values(by=['userID'])
book
```

```python
book_df = book.pivot_table(index='userID', columns='booktitle',values='bookrating').reset_index(drop=True)
book_df
```

```python
book_df.fillna(0, inplace=True)
```

```python
book_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_book = 1 - pairwise_distances(book_df.values,metric='cosine')
```

```python
user_book
```

```python
user_book_df = pd.DataFrame(user_book)
```
